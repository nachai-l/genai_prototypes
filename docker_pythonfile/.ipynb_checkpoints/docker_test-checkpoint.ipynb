```python
from pathlib import Path
import docker, json
from docker.errors import BuildError, APIError

def print_build_logs(logs):
    for chunk in logs:
        line = chunk.get("stream") or chunk.get("status") or chunk.get("error") or chunk.get("aux")
        if line is None:
            line = json.dumps(chunk)
        if isinstance(line, (bytes, bytearray)):
            line = line.decode("utf-8", "ignore")
        print(line, end="")

def build_image(
    context_dir: str | Path,
    tag: str,
    dockerfile: str = "Dockerfile",   # relative to context_dir
    platform: str | None = None,      # e.g. "linux/amd64" on Apple Silicon
    buildargs: dict | None = None,    # e.g. {"TARGET": "prod"}
    target: str | None = None,        # multi-stage target
    nocache: bool = False,
    pull: bool = True,                # like `--pull`
    labels: dict | None = None,
    network_mode: str | None = None,  # e.g. "host" (Linux)
):
    client = docker.from_env()
    try:
        image, logs = client.images.build(
            path=str(Path(context_dir).resolve()),  # <<< build context
            tag=tag,
            dockerfile=dockerfile,
            platform=platform,
            buildargs=buildargs or {},
            target=target,
            nocache=nocache,
            pull=pull,
            rm=True,
            labels=labels,
            network_mode=network_mode,
        )
        print_build_logs(logs)
        return image
    except BuildError as e:
        print("Build failed. Last logs:")
        print_build_logs(e.build_log or [])
        raise
    except APIError:
        raise
```

```python
# === Examples ===
# 1) equivalent to: docker build -t myapp:ep path/to/myapp
img = build_image("myapp", tag="myapp:ep", dockerfile="deploy/Dockerfile.prod")
```

```python
# 2) custom Dockerfile inside the same context (relative path):
#    docker build -f deploy/Dockerfile.prod -t myapp:prod path/to/myapp
img = build_image(
    "myapp",
    tag="myapp:prod",
    dockerfile="deploy/Dockerfile.prod",
)
```

```python
# 3) use build args + select a multi-stage target, no cache, pull base each time
img = build_image(
    "myapp",
    tag="myapp:prod",
    buildargs={"TARGET": "prod"},
    target="runtime",
    nocache=True,
    pull=True,
    dockerfile="deploy/Dockerfile.prod"
)
```

```python
import docker
client = docker.from_env()

cfg = client.images.get("myapp:prod").attrs["Config"]
print("Entrypoint:", cfg.get("Entrypoint"))
print("Cmd:", cfg.get("Cmd"))
```

```python
# If your image has ENTRYPOINT ["python","-u","app.py"]
print(client.containers.run(
    "myapp:prod",
    command=["hello", "--name", "Kelvin"],
    detach=False, remove=True
).decode())
```

```python
ct = client.containers.run("myapp:prod", detach=True, name="myapp_run")
ct.wait()  # block until exit (or skip for true daemons)
print(ct.logs(stdout=True, stderr=True).decode())
ct.remove()
```

```python
# env + port map + volume mount
out = client.containers.run(
    "myapp:prod",
    environment={"APP_ENV": "prod"},
    ports={"8000/tcp": 8000},             # container:host
    volumes={"/host/data": {"bind": "/app/data", "mode": "ro"}},
    detach=False, remove=True
)
print(out.decode())
```
